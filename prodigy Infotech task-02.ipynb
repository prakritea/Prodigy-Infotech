```python
data = data.drop(columns=['Cabin'])
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
data
```

```python
data['Embarked']=data['Embarked'].fillna(0)
```

```python
data.duplicated().sum()
```

```python
sns.countplot(x='Survived',data=data,palette='husl',)
plt.xlabel("Survival")
plt.ylabel("Number of people")
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.show()
```

```python
sns.countplot(x='Sex',data=data,palette='husl',)
plt.xlabel("Gender")
plt.ylabel("Number of people")
plt.show()
```

```python
sns.countplot(x='Survived',hue='Sex',data=data,palette='husl',)
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.show()
```

```python
sns.countplot(x='Pclass',data=data,palette='husl',)
plt.xlabel("Pclass")
plt.ylabel("Number of people")
plt.show()
```

```python
sns.catplot(x = 'Pclass', hue = 'Survived', col = 'Sex', kind = 'count', data =
data,palette='husl' )
plt.tight_layout()
```

```python
sns.countplot(x='Embarked',data=data,palette='coolwarm',)
plt.xlabel("Embarked")
plt.ylabel("Number of people")
plt.show()
```

```python
fig,axes = plt.subplots(1, 2, figsize=(12, 6))
sns.countplot(x='SibSp',data=data,ax=axes[0],palette='husl')
sns.countplot(x='Parch',data=data,ax=axes[1],palette='husl')
plt.show()
```

```python
sns.countplot(x ='Survived', hue='SibSp',data=data,palette='viridis')
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.title("Survival population of Sibsp")
plt.show()
```

```python
sns.countplot(x ='Survived',hue='Parch',data=data,palette='viridis')
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.title("Survival population of Parch")
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.show()
```

```python
fig,axes = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(data['Fare'], kde=True,ax=axes[0])
sns.histplot(data['Age'].dropna(),kde=True,ax=axes[1])
plt.show()
```
